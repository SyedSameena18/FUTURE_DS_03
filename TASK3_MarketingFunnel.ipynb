## KPI METRICS

```python
#TOTAL IMPRESSIONS KPI
total_impressions = df['Impressions'].sum()
print("Total Impressions:", total_impressions)
```

```python
#TOTAL CLICKS KPI
total_clicks = df['Clicks'].sum()
print("Total Clicks:", total_clicks)
```

```python
#AVERAGE CONVERSION RATE
avg_conversion = df['Conversion_Rate'].mean()
print("Average Conversion Rate:", round(avg_conversion, 2))
```

# DATA VISUALIZATION

```python
#Channel Performance by Conversion
channel_performance = df.groupby('Channel_Used')['Conversion_Rate'].mean().sort_values()
channel_performance.plot(kind='bar')
plt.title("Channel Performance by Conversion Rate")
plt.ylabel("Average Conversion Rate")
plt.show()
```

```python
# Impressions vs Clicks
plt.scatter(df['Impressions'], df['Clicks'])
plt.title("Impressions vs Clicks")
plt.xlabel("Impressions")
plt.ylabel("Clicks")
plt.show()
```

```python
#ROI ANALYSIS VISUALIZATION
roi_analysis = df.groupby('Channel_Used')['ROI'].mean().sort_values()
roi_analysis.plot(kind='bar')
plt.title("Average ROI by Marketing Channel")
plt.xlabel("Marketing Channel")
plt.ylabel("ROI")
plt.show()
```

```python
#CUSTOMER SEGMENT ANALYSIS
segment_analysis = df.groupby('Customer_Segment')['Conversion_Rate'].mean().sort_values()
segment_analysis.plot(kind='bar')
plt.title("Conversion Rate by Customer Segment")
plt.xlabel("Customer Segment")
plt.ylabel("Conversion Rate")
plt.show()
```
